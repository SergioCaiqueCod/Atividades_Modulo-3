```python
# FUNÇÕES SÃO PARA TAREFAS QUE SE REPETEM
def calcula_media_lista(lista):
    soma =sum(lista)#somo todos os valores
    quantidade= len(lista) # pego a quantidade de elementos
    media = soma/quantidade #calculo a media
    return media #retorno a media

media_da_lista = calcula_media_lista([1,2,3,4,5])
print(media_da_lista)
```

```python
def conta_vogais(texto):
    contador = 0
    Vogais=('a','e','i','o','u')
    texto = texto.lower() # transforma as letras em minusculos e UPPER transforma em maiusculo
    for letra in texto: # percorro as letras do texto
        if letra in Vogais: # se a letra é uma vogal
            contador += 1 # conta mais um
   return contador     
print(conta_vogais(input('digite seu nome:')))
```
