```python
import numpy as np  # Importa a biblioteca NumPy, que é usada para trabalhar com arrays e operações numéricas

# Gera um array de 10 números inteiros aleatórios entre 0 e 100 (inclusive)
# np.random.randint(início, fim_exclusivo, tamanho)
array = np.random.randint(0, 101, 10)

# Calcula a média dos valores do array
# np.mean() retorna a média aritmética dos elementos
media = np.mean(array)

# Filtra apenas os valores maiores que a média
# A expressão (array > media) cria um array booleano
# Usar isso como índice retorna apenas os elementos que satisfazem a condição
valores_maiores = array[array > media]

# Exibe o array original
print("Array original:", array)

# Exibe a média calculada
print("Média:", media)

# Exibe os valores filtrados
print("Valores maiores que a média:", valores_maiores)
```

```python
dados=np.array([5, 7, 8, 6, 4, 9, 10])

media=np.mean(dados) #calcular a media

print(np.sum(dados > media)) # conta quantos elementos estão acima da média, já que verdadeiro vale 1 e falso 0
print(len(dados[dados>media])) # Aqui ele está contandos elementos maiores que a media...


```
