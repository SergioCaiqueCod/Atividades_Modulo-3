```python
nota1=int(input("Sua nota:"))
nota2=int(input("Sua nota:"))
nota3=int(input("Sua nota:"))

conta=(nota1+nota2+nota3 )
resultado=(conta/3)
print(resultado)

if resultado >= 7:
    print('Aprovado')
    
elif resultado >= 5 and resultado < 7:
    print('Recuperação')

elif resultado < 5:
    print('Reprovado')

```
